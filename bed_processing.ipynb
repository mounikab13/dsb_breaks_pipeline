```python
import pandas as pd

def read_bed_file(bed_file_path):
    """
    Read a BED file into a DataFrame.

    Parameters:
    - bed_file_path (str): The path to the BED file.

    Returns:
    - pd.DataFrame: The DataFrame containing the BED file data.
    """
    columns = ["chromosome", "start", "end"]
    bed_df = pd.read_csv(bed_file_path, sep="\t", header=None, names=columns)
    return bed_df

```

```python
def count_rows_in_dataframes(dataframes, sample_names):
    """
    Count the number of rows in a list of DataFrames and create a summary DataFrame.

    Parameters:
    - dataframes (list of pd.DataFrames): List of DataFrames to count rows.
    - sample_names (list of str): List of sample names corresponding to the DataFrames.

    Returns:
    - pd.DataFrame: Summary DataFrame with sample names and row counts.
    """
    summary_data = {'sample': sample_names, 'sample_breaks': [df.shape[0] for df in dataframes]}
    summary_df = pd.DataFrame(summary_data)
    return summary_df


### actual execution
sample_dataframes = [sample1, sample2, sample3, sample4, sample5, sample6, sample7, sample8, sample9, sample10, sample11, sample12,
                     sample13, sample14, sample15, sample16]

sample_names = total_breaks["sample"]

count_rows = [df.shape[0] for df in sample_dataframes]

summary_dataframe = pd.DataFrame({'sample': sample_names, 'dsb_count': count_rows})

summary_dataframe
```

```python
merged_df = pd.merge(total_breaks, summary_dataframe, on='sample', how='inner')

merged_df
```

```python
merged_df["normalized_breaks"] = merged_df["dsb_count"]/(merged_df["total_breaks"]/1000)

merged_df
```
